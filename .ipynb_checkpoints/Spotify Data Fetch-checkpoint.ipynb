```python
import requests
import base64
import pandas as pd

def get_token(client_id, client_secret):
    """
    Authenticates with the Spotify API and returns an access token.
    """
    auth_url = 'https://accounts.spotify.com/api/token'
    auth_header = base64.b64encode(f"{client_id}:{client_secret}".encode()).decode('utf-8')
    auth_payload = {'grant_type': 'client_credentials'}
    auth_headers = {'Authorization': f'Basic {auth_header}'}

    try:
        response = requests.post(auth_url, data=auth_payload, headers=auth_headers)
        response.raise_for_status()
        json_response = response.json()
        return json_response.get('access_token')
    except requests.exceptions.RequestException as e:
        print(f"Error during authentication: {e}")
        return None

def get_popular_tracks(token, playlist_id=None):
    """
    Fetches the tracks from a playlist and returns a DataFrame.
    Includes Track Name, Artists, Album, and Popularity.
    """
    if not playlist_id:
        print("Please provide a valid playlist_id.")
        return pd.DataFrame()

    tracks_url = f'https://api.spotify.com/v1/playlists/{playlist_id}/tracks'
    headers = {'Authorization': f'Bearer {token}'}

    try:
        tracks_response = requests.get(tracks_url, headers=headers)
        tracks_response.raise_for_status()
        tracks_data = tracks_response.json()

        tracks = tracks_data.get('items', [])
        track_list = []

        for track_item in tracks:
            track = track_item.get('track', {})
            if not isinstance(track, dict):
                continue
            track_id = track.get('id')
            track_name = track.get('name')
            artists = [artist['name'] for artist in track.get('artists', [])]
            album_name = track.get('album', {}).get('name')
            popularity = track.get('popularity')  # ðŸŽ¯ Add popularity score

            if track_name and artists:
                track_list.append({
                    "Track Name": track_name,
                    "Track ID":track_id,
                    "Artists": ", ".join(artists),
                    "Album": album_name,
                    "Popularity": popularity
                })

        return pd.DataFrame(track_list)

    except Exception as e:
        print(f"Error fetching tracks: {e}")
        return pd.DataFrame()

def main(client_id, client_secret):
    """
    Main function to run the script.
    """
    token = get_token(client_id, client_secret)
    if token:
        playlist_id = "3jBzuxVcBq3AJAYGDKyhgz"  # Example playlist
        df = get_popular_tracks(token, playlist_id)
        return df
    else:
        print("Failed to get token.")
        return pd.DataFrame()


# Example usage:
# client_id = "YOUR_CLIENT_ID"
# client_secret = "YOUR_CLIENT_SECRET"
# df = main(client_id, client_secret)
# print(df.head())
```
